## Let`s define Input and output
## X is the input.
## y is the output

```python
X=df.drop(['Class'], axis = 1).values 
y= df['Class'] 
```

```python
print(X)
```

```python
print(y)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

#  Feature extraction step
First, we will use Selectpercintile

```python
# Select features according to a percentile of the highest scores.
from sklearn.feature_selection import SelectPercentile
sp=SelectPercentile(percentile=15)
# p referes to perecentile
Xp=sp.fit_transform(X,y)
```

```python
print(Xp.shape)
```

```python
# Train test split
# sp referes to selectperecentile
from sklearn.model_selection import train_test_split
X_train_sp,X_test_sp,y_train_sp,y_test_sp=train_test_split(Xp,y,test_size=0.25,random_state=42)
```

```python
print(" The shape of X_train in select_percentile is ",X_train_sp.shape)
print(" The shape of X_test in select_percentile is ",X_test_sp.shape)
print(" The shape of y_train in select_percentile is ",y_train_sp.shape)
print(" The shape of y_test in select_percentile is ",y_test_sp.shape)
```

second : We will use SelectKBest.

```python
# b referecs to best
from sklearn.feature_selection import SelectKBest
kb=SelectKBest(k=5)
Xb=kb.fit_transform(X,y)
```

```python
print(Xb.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train_kb,X_test_kb,y_train_kb,y_test_kb=train_test_split(Xb,y,test_size=0.25,random_state=42)
```

```python
print(" The shape of X_train_kb in Kbest is ",X_train_kb.shape)
print(" The shape of X_test_kb in Kbest is ",X_test_kb.shape)
print(" The shape of y_train_kb in Kbest is ",y_train_kb.shape)
print(" The shape of y_test_kb in Kbest is ",y_test_kb.shape)
```

Third : We will use GenericUnivariateSelect

```python
# G refers to generic
from sklearn.feature_selection import GenericUnivariateSelect
gu=GenericUnivariateSelect(param=14)
Xg=gu.fit_transform(X,y)
```

```python
print(Xg.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train_g,X_test_g,y_train_g,y_test_g=train_test_split(Xg,y,test_size=0.25,random_state=42)
```

```python
print(" The shape of X_train in GenericUnivariateSelect is ",X_train_g.shape)
print(" The shape of X_test in GenericUnivariateSelect is ",X_test_g.shape)
print(" The shape of y_train in GenericUnivariateSelect is ",y_train_g.shape)
print(" The shape of y_test in GenericUnivariateSelect is ",y_test_g.shape)
```

Fourth : We will use SelectFpr

```python
from sklearn.feature_selection import SelectFpr
# the number of zeroes in alpha refers to the number of fearures that will be reduced
sl=SelectFpr(alpha=0.000001)
# s refers to select
Xs=sl.fit_transform(X,y)
```

```python
print(Xs.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train_s,X_test_s,y_train_s,y_test_s=train_test_split(Xs,y,test_size=0.25,random_state=42)
```

```python
print(" The shape of X_train in SelectFpr is ",X_train_s.shape)
print(" The shape of X_test in SelectFpr is ",X_test_s.shape)
print(" The shape of y_train in SelectFpr is ",y_train_s.shape)
print(" The shape of y_test in SelectFpr is ",y_test_s.shape)
```

## 1- Under Sampling¶


## Selectpercintile

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
from imblearn.under_sampling import NearMiss
ns = NearMiss(0.7)
X_train_ns_sp, Y_train_ns_sp = ns.fit_sample(X_train_sp,y_train_sp)
X_test_ns_sp, Y_test_ns_sp = ns.fit_sample(X_test_sp,y_test_sp)
```

```python
print(df.Class.value_counts())
```

```python
Y_train_ns_sp.value_counts()
```

```python
rfc_sc=RandomForestClassifier(random_state=0)
rfc_sc.fit(X_train_ns_sp,Y_train_ns_sp)
```

```python
Y_pred = rfc_sc.predict(X_test_ns_sp)
print(confusion_matrix(Y_test_ns_sp, Y_pred))
print(accuracy_score(Y_test_ns_sp,Y_pred))
print(classification_report(Y_test_ns_sp,Y_pred))
```
